```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow as tf
from skimage.transform import resize
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.optimizers import SGD
from keras import Model
import cv2
import os
```

```python
path = 'C:/Users/n10941576/OneDrive - Queensland University of Technology/python(vscode)/CNN/dataset'
```

```python
os.listdir(path)
image_file_path = [path + '/' + i for i in os.listdir(path)]
```

```python
data = np.empty((1000, 224, 224, 3))
index = 0
for i in image_file_path:
    image_path = [i + '/' + j for j in os.listdir(i)]
    for loc in image_path:
        image = resize(cv2.imread(loc), output_shape=(224, 224, 3))
        data[index] = image
        index += 1
```

```python
label = np.ndarray(shape=(1000,5), dtype = int)
for i in range(200):
  label[i] = [1, 0, 0, 0, 0]
  label[i+200] = [0, 1, 0, 0, 0]
  label[i+400] = [0, 0, 1, 0, 0]
  label[i+600] = [0, 0, 0, 1, 0]
  label[i+800] = [0, 0, 0, 0, 1]
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, label, test_size = 0.2)
```

```python
model = tf.keras.applications.mobilenet_v2.MobileNetV2(weights = 'imagenet')
```

```python
output = Dense(5, activation='softmax')
output = output(model.layers[-2].output)
Input = model.input
model = Model(inputs = Input, outputs = output)
```

```python
model.summary()
```

```python
optimizerSGD = tf.keras.optimizers.SGD(learning_rate = 0.01, momentum = 0.0, nesterov = False)
model.compile(loss = "categorical_crossentropy", optimizer = optimizerSGD, metrics = ["accuracy"])
```

```python
model.fit(x_train, y_train, validation_split = 0.2, epochs = 50, verbose = 2)
```
